```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime
from scipy.stats import linregress
import scipy.stats as st
```

```python
date_now = datetime.date(datetime.now())

def drawRegressionPlots(x_values, y_values, xlabel, ylabel, title, postion):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    
    print(f'slope : {slope}')
    print(f'intercept : {intercept}')
    print(f'rvalue : {rvalue}')
    print(f'pvalue : {pvalue}')
    print(f'stderr : {stderr}')

    regress_values = x_values * slope + intercept
    print(f'regress_values: {regress_values}')

    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    print(f'line_eq: {line_eq}')

    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r")
    plt.annotate(line_eq,(1,5),fontsize=19,color="red")
#     plt.xlabel('Rooms in House')
    plt.xlabel(xlabel, fontsize = 3)
#     plt.ylabel('Median House Prices ($100,000)')
    plt.ylabel(ylabel, fontsize = 3)
#     plt.title(f"{hemisphere} Hemisphere - {x_label} vs {y_label} {date_now}",fontsize = 15)

    plt.title(title + " "+ {date_now}, fontsize= 10)
#     plt.title(f"abc",fontsize = 15)
    plt.show()

     # Print r square value
    print(f"The r-squared is: {rvalue**2}")
    correlation = st.pearsonr(x_values,y_values)
    print(f"The correlation between both factors is {round(correlation[0],2)}")
    
    
    
    
    


# x_values = np.array([1, 3, 5, 7])
# y_values = np.array([ 10, 30, 30, 70 ])

# (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

# print(f'slope : {slope}')
# print(f'intercept : {intercept}')
# print(f'rvalue : {rvalue}')
# print(f'pvalue : {pvalue}')
# print(f'stderr : {stderr}')

# regress_values = x_values * slope + intercept
# print(f'regress_values: {regress_values}')

# line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# print(f'line_eq: {line_eq}')

# plt.scatter(x_values,y_values)
# plt.plot(x_values,regress_values,"r")
# plt.annotate(line_eq,(1,5),fontsize=19,color="red")
# plt.xlabel('Rooms in House')
# plt.ylabel('Median House Prices ($100,000)')
# # plt.title(f"{hemisphere} Hemisphere - {x_label} vs {y_label} {date_now}",fontsize = 15)
# plt.title(f"abc",fontsize = 15)
# plt.show()

#  # Print r square value
# print(f"The r-squared is: {rvalue**2}")
# correlation = st.pearsonr(x_values,y_values)
# print(f"The correlation between both factors is {round(correlation[0],2)}")
```
